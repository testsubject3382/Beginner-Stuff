```python
def caesar():
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    shift = 5
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet, shifted_alphabet)
    text = 'hello world'
    encrypted_text = text.translate(translation_table)
    print(encrypted_text)
```

```python
#Beginner stuff Alison 2

#List
list1=['apple','banana','orange','mango','pear','pineapple']
list1[3]='horse'
print(list1[::-1])
#add
list1.append('kiwi')
print(list1)
#remove
list1.remove('kiwi')
print(list1)
list1.pop(4)
print(len(list1))

#Bubble sort
my_list=[8,3,1,4] # Renamed 'list' to 'my_list' to avoid shadowing built-in type
length=len(my_list)
print(my_list)
for i in range(length):
  for j in range(0,length-i-1): # Changed inner loop variable from 'i' to 'j'
    if my_list[j]>my_list[j+1]: # Used 'j' for indexing
      my_list[j],my_list[j+1]=my_list[j+1],my_list[j]
print(my_list)
```

```python
# Scopes (Global, Local-inside the function)

x=10  #Global
def function():
  x=20  #Local
function()
print(x)

def function1():
  global x  #Making a local function into a global function
  x=20
function1()
print(x)
```
